```python
import torch
```

Indexing

```python
tensor = torch.tensor([
                       [10,20,30],
                       [40, 50, 60],
                       [70, 80, 90]])
print(tensor.ndim)
print(tensor.shape)
```

```python
# Access first row
tensor[0]
```

```python
# Access first column
tensor[:,0]
```

```python
# First Twor rows
print("First two rows : \n",tensor[:2])
```

```python
# Middle elements (row 1, column 1 and 2 only)
print(tensor[1,1:3])
```

```python
# Fancy indexing
# (0,1)---> 20
# (2,2) ---> 90

selected_elements = tensor[[0,2],[1,2]]
print(selected_elements)
```

```python
# boolean indexing
mask = tensor>50
print(mask)
print(tensor[mask])
```

```python
# Changingtensor values via indexing
print(tensor)
tensor[0,1]=25
print(tensor)
tensor[0,0]=786
print(tensor)
```

```python
tensor[:,0] = torch.tensor([100,200,300])
print(tensor)
```

```python
indices = torch.tensor([0,2])
selected_rows = torch.index_select(tensor,dim=1,index=indices)
print(selected_rows)
```

```python
print(tensor[:,::2])
```

```python
torch.flip(tensor,dims=(0,))
```
